```python
# historical data
def bucketed(time_bucket):
    df = pd.read_csv('./data/fake_news_tweets.csv')
    df['day_of_week'] = pd.to_datetime(df['created_at']).dt.day_name()
    
    df['created_at'] = pd.to_datetime(df['created_at']).dt.date
    yesterday = datetime.date.today() - datetime.timedelta(1)
    tweet_counts = df.groupby('created_at').agg('count')["text"]
    dates = pd.date_range('2017-01-25', yesterday, freq='D')
    tweet_counts
    counts = pd.DataFrame({ "count": tweet_counts},index=dates).fillna(0)
    return counts.resample(time_bucket).sum()
```

```python
# number of fake news tweets per week
df["week_of"] = pd.to_datetime(df['created_at']) - pd.to_timedelta(7, unit='d')
tweets_per_week = df.groupby([pd.Grouper(key='week_of', freq='W-MON')]).size()
tweets_per_week.columns = ["n"]
tweets_per_week.plot()
```

```python
import requests
import json
import pprint

## external data
positions = { 
    0: {
        "yes": [(.35, 10)]
    },
    1: {
        "yes": [(.22, 15)]
    },
    2: {
        "yes": [(.16, 5), (.14, 12)]
    },
    4: {
        "no": [(.92, 16), (.94, 2)]
    },
    5: {
        "no": [(.96, 2)]
    },
    9: {
        "no": [(.97, 4)]    
    }
}

ACCOUNT_BALANCE = 100

## value functions
def expected_value(potential_win, chance_win, potential_loss, chance_loss):
    return (potential_win * chance_win) - (potential_loss * chance_loss)

def allocation(account_balance, expected_value):
    return account_balance * (( expected_value * 5 ) / 10)

def recommended_shares(account_balance, expected_value, price_per_share):
    return allocation(account_balance, expected_value) / price_per_share;

def to_proba(buckets):
    vals = buckets.value_counts()
    s = vals.sum()
    return vals/s

## portfolio management
def shares_bought(c, positions):
    bought = 0
    if c in positions:
        for pos in positions[c]:
            bought += pos[1]
    return bought

def recommend_buy(contract, account_balance, expected_value, price_per_share):
    shares = recommended_shares(account_balance, expected_value, price_per_share) - shares_bought(contract)
    print("BUY shares for contract {n}: {shares} shares @{price}".format(n=contract,shares=shares, price=price_per_share))

def recommend_sell(contract, n_shares, price_per_share):
    print("SELL shares for contract {n}: {shares} shares @{price}".format(n=contract,shares=n_shares, price=price_per_share))

## market evaluation
def fetch_market_data(market_id):
    url = "https://www.predictit.org/api/marketdata/markets/{id}".format(id=market_id)
    r = requests.get(url=url)
    return r.json()

def eval_market(market_id):
    data = fetch_market_data(market_id)
    contracts = data["contracts"]
    
    n_days = days_left(contracts[0]["dateEnd"])
    b=bucketed("{n}D".format(n=n_days))
    proba = to_proba(b['count'])
    pprint.pprint(proba)
    n_tweets = 0
    
    for c in contracts: 
        expected_values = pprint.pprint(eval_trade_variations(c, proba, n_tweets))
        pprint.pprint(expected_values)
        for k, v in expected_values:
            if v > 0:
                if "buy" in k:
                    recommend_buy(c, v)

def eval_trade_variations(contract, proba, curr_n_tweets):
    _n_tweets = int( contract["name"].split(' ')[0] )
    n_tweets = _n_tweets - curr_n_tweets 
    if n_tweets not in proba: 
        proba_yes = 0
    else:
        proba_yes = proba[n_tweets]
    
    buy_yes = contract["bestBuyYesCost"]
    buy_no = contract["bestBuyNoCost"]
    sell_yes = contract["bestSellYesCost"]
    sell_no = contract["bestSellNoCost"]
    
    if buy_yes and buy_no and 1 - buy_yes - buy_no > 0:
            print("Arbitrage opportunity BUY:", _n_tweets, "contract, ", buy_yes, buy_no)
    
    d = {}
    d[n_tweets] = {}
    
    if buy_yes:
        d[n_tweets]["buy_yes"] = (expected_value(1-buy_yes, proba_yes, buy_yes, 1-proba_yes), buy_yes)
        
    if buy_no:
        d[n_tweets]["buy_no"] = (expected_value(1-buy_no, 1-proba_yes, buy_no, proba_yes), buy_no)
            
    if n_tweets in positions and "yes" in positions[n_tweets]:
        yes_positions = positions[n_tweets]["yes"]
        for pos in yes_positions:
            strike_price = pos[0]
            quantity = pos[1]
            ev = (sell_yes - strike_price) - expected_value(1-strike_price, proba_yes, strike_price, 1-proba_yes)
            key = "sell_yes_"+str(strike_price)+"_"+str(quantity)
            d[n_tweets][key] = (ev, sell_yes)
            
    if n_tweets in positions and "no" in positions[n_tweets]:
        no_positions = positions[n_tweets]["no"]
        for pos in no_positions:
            strike_price = pos[0]
            quantity = pos[1]
            ev = (sell_no - strike_price) - expected_value(1-strike_price, 1-proba_yes, strike_price, proba_yes)
            key = "sell_no_"+str(strike_price)+"_"+str(quantity)
            d[n_tweets][key] = (ev, sell_no)
    return d

def days_left(end_date_str):
    end_date = parser.parse(end_date_str)
    start_date = end_date - datetime.timedelta(days=7)
    delta = datetime.datetime.now() - start_date
    days_left = ((7*24) - (delta.total_seconds()/3600))/24
    return round(days_left)

eval_market(5371)
```
